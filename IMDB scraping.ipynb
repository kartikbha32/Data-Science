```python
from bs4 import BeautifulSoup
```

```python
import requests 
```

```python
response = requests.get('https://www.imdb.com/chart/top')
```

```python
soup = BeautifulSoup(response.text,'lxml')
```

```python
movie_name = soup.find_all(class_='titleColumn')
```

```python
yearofrelease = soup.find_all(class_='secondaryInfo')
```

```python
rating = soup.find_all(class_='ratingColumn imdbRating')
```

```python
rating
```

```python
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
```

```python
links
```

```python
l=[]
```

```python
for i in links:
    response = requests.get('http://www.imdb.com'+i)
    soup = BeautifulSoup(response.text,'lxml')
    director_name = soup.find(class_='credit_summary_item')
    inner_text = director_name.text
    strings = inner_text.split("\n")
    c=strings[2]
    l.append(c)
    
        
```
